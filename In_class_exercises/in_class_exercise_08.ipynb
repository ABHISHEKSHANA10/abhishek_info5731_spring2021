```python
import re
from nltk.corpus import stopwords
from textblob import Word
import pandas as pd
import nltk
```

```python
def special_char_count(given_input):
  special_char = 0
  for i in range(len(given_input)):
    if not((given_input[i].isalpha()) and (given_input[i].isdigit())):
        special_char = special_char + 1
  return special_char
```

```python
#special characters removal:

data['Removal of special characters'] = data['Reviews'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))

# punctuations removal:

data['Punctuation'] = data['Removal of special characters'].str.replace('[^\w\s]','')

#Removal of numbers:

data['Removal of numbers'] = data['Punctuation'].str.replace('\d+', '')

#Lower case all texts:

data['Lower Case words'] = data['Removal of numbers'].apply(lambda x: " ".join(x.lower() for x in x.split()))

#Removing stopwords by using stopwordslist

stop = stopwords.words('english')
data['removed stopwords'] = data['Lower Case words'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
data
```

```python
from textblob import TextBlob
words = []
sentimental_words = []
for i in data['removed stopwords']:
  words.append(i.split(' '))
for i in words:
  for j in i:
    if TextBlob(j).sentiment.polarity != 0:
      sentimental_words.append(j)
print(sentimental_words)
```

```python
from collections import Counter
import numpy as np
sentimental_word_count = Counter(sentimental_words)
df_words = pd.DataFrame(list(zip(list (sentimental_word_count.keys()),list (sentimental_word_count.values()))), columns = ['words', 'Frequency'])
df_words = df_words.sort_values(by=['Frequency'], ascending=False)
df_words.insert(0, "Rank", np.arange(1,len(df_words)+1), True)
df_words
```

```python
original_positive = 0
original_negative = 0
original_neutral = 0
for value in data['Sentiment']:
  if value == 'Positive':
    original_positive += 1
  elif value == 'Negative':
    original_negative += 1
  elif value == 'Neutral':
    original_neutral += 1
print('Original Positive = ', original_positive)
print('Original Negative = ', original_negative)
print('Original Neutral = ', original_neutral)
```

```python
from textblob import TextBlob
positive = 0
negative = 0
neutral = 0
for i in data['removed stopwords']:
  sent = TextBlob(i)
  if sent.sentiment.polarity > 0:
    positive += 1
  elif sent.sentiment.polarity < 0:
    negative += 1
  elif sent.sentiment.polarity == 0:
    neutral += 1
print("Positive = ", positive)
print("Negative = ", negative)
print("Neutral = ", neutral)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
positive = 0
negative = 0
neutral = 0
sentimental = SentimentIntensityAnalyzer()
for i in data['removed stopwords']:
  sent = TextBlob(i)
  if sentimental.polarity_scores(i)['compound'] > 0:
    positive += 1
  elif sentimental.polarity_scores(i)['compound'] < 0:
    negative += 1
  elif sentimental.polarity_scores(i)['compound'] == 0:
    neutral += 1
print("Positive = ", positive)
print("Negative = ", negative)
print("Neutral = ", neutral)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import classification_report
vectorizer = TfidfVectorizer(min_df = 5,
                             max_df = 0.8,
                             sublinear_tf = True,
                             use_idf = True)
train_1, test_1, train_2, test_2 = train_test_split(data['removed stopwords'],data['Sentiment'],test_size=0.33)
train = vectorizer.fit_transform(train_1)
test = vectorizer.transform(test_1)
linear_class = svm.SVC(kernel='linear')
linear_class.fit(train, train_2)
linear_pred = linear_class.predict(test)
output = classification_report(test_2, linear_pred, output_dict=True)
print('positive: ', output['Positive'])
print('negative: ', output['Negative'])
print("neutral:", output['Neutral'])
```
