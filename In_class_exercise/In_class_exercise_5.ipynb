```
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from textblob import Word
import nltk
from nltk.tokenize import word_tokenize
```

```
content = []
first_page = 'https://citeseerx.ist.psu.edu/search;jsessionid=37A0AC54277865A394D5F96748080D17?q=natural+language+processing&t=doc&sort=date'
for pg_no in range(50):
  if pg_no == 0:
    url = first_page
  else:
    url = 'https://citeseerx.ist.psu.edu/search?q=natural+language+processing&t=doc&sort=date&start=' + str (pg_no)
  page = requests.get(url)
  soup = BeautifulSoup(page.content, 'html.parser')
  abstracts = soup.find_all(class_='pubabstract')
  pg_no = pg_no + 10
  for abstract in abstracts:
    processed_text = abstract.text.replace('\n', '').strip()
    content.append(processed_text)
df = pd.DataFrame((content), columns =['abstract'])
df
```

```
import nltk
nltk.download()
```

```
#lower casing
df['lower casing'] = df['abstract'].apply(lambda x: " ".join(x.lower() for x in x.split()))
#special chars removal
df['Special Characters Removal'] = df['abstract'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', i) for i in x ))
#Removal of StopWords
stop = stopwords.words('english')
df['removed stopwords'] = df['abstract'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))

#Stemming Data
st = PorterStemmer()
df['Stemming'] = df['removed stopwords'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
#Lemmatization Data
nltk.download('wordnet')
df['Lemmatization'] = df['Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df
```

**Parts of speech and tagging**

```
parts_of_speech = []
for sentence in df['lower casing']:
  parts_of_speech.append(nltk.pos_tag(word_tokenize(sentence)))
print(parts_of_speech)
```

```
counts = {}
result = {}
result['Cordinating Conjuction'] = []
result['Determiner'] = []
result['Pre determiner'] = []
result['Modal'] = []
result['subordinating conjuction'] = []
result['Adjective'] = []
result['Proper Noun'] = []
result['Verb'] = []
result['Possessive Ending'] = []
result['Possesive Pronoun'] = []

for i in parts_of_speech:
  counts['CC'] = counts['DT'] = counts['PDT'] = counts['MD'] = counts['IN'] = counts['JJ'] = counts['NNP'] = counts['VB'] = counts['POS'] = counts['PRP$'] = 0
  for j in i:
    if j[1] == 'CC':
      counts['CC'] += 1
    elif j[1] == 'DT':
      counts['DT'] += 1
    elif j[1]== 'PDT':
      counts['PDT'] +=1
    elif j[1] == 'MD':
      counts['MD'] += 1
    elif j[1] == 'IN':
      counts['IN'] += 1
    elif j[1] == 'JJ':
      counts['JJ'] += 1
    elif j[1] == 'NNP':
      counts['NNP'] += 1
    elif j[1] == 'VB':
      counts['VB'] += 1
    elif j[1] == 'POS':
      counts['POS'] += 1
    elif j[1] == 'PRP$':
      counts['PRP$'] += 1
  result['Cordinating Conjuction'].append(counts['CC'])
  result['Determiner'].append(counts['DT'])
  result['Pre determiner'].append(counts['PDT'])
  result['Modal'].append(counts['MD'])
  result['subordinating conjuction'].append(counts['IN'])
  result['Adjective'].append(counts['JJ'])
  result['Proper Noun'].append(counts['NNP'])
  result['Verb'].append(counts['VB'])
  result['Possessive Ending'].append(counts['POS'])
  result['Possesive Pronoun'].append(counts['PRP$'])
```

```
dataframe = pd.DataFrame.from_dict(result)
dataframe.insert (0, "abstract", df['lower casing'])
dataframe
```
