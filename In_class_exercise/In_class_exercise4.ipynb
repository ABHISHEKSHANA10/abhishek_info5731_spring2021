```
from six.moves import urllib
import pandas as pd

text = []
url = "https://raw.githubusercontent.com/unt-iialab/info5731_spring2021/main/class_exercises/01-05-1%20%20Adams%20v%20Tanner.txt"
data = urllib.request.urlopen(url)

for line in data:
  cleandata = line.decode("utf-8").replace('\r\n', '')
  if cleandata: text.append(cleandata)
df = pd.DataFrame (text, columns = ['content'])
print(df)
```

```
#No.of sentences
df['Number of Sentences'] = df['content'].apply(lambda x: len(str(x).split(".")))
df
```

```
#No.of words
df['word count'] = df['content'].apply(lambda x: len(str(x).split(" ")))
df
```

```
#No.of characters
df['Characters_count'] = df['content'].str.len()
df
```

```
#Average word length
def avg(sent):
  words = sent.split()
  if len(words) != 0:
    return(sum(len(value) for value in words)/len(words))
  else:
    return None

df['Average Length of words'] = df['content'].apply(lambda x: avg(x))
df
```

```
#No.of stopwords
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop = stopwords.words('english')

df['Stopwords count'] = df['content'].apply(lambda x: len([x for x in x.split() if x in stop]))
df
```

```
#no. of special chars
def special_characters_count(sent):
  special_count = 0
  for i in range(len(sent)):
    if not((sent[i].isalpha()) and (sent[i].isdigit())):
        special_count = special_count + 1
  return special_count

df['No.of special characters'] = df['content'].apply(lambda x: special_characters_count(x))
df
```

```
#No.of numerics
df['No.of numerics'] = df['content'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
df
```

```
df['upper case words count'] = df['content'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
df
```

```
#1.2
#Lower casing
df['Lower Casing'] = df['content'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df
```

```
#punctuation removal
df['Punctuation removal'] = df['Lower Casing'].str.replace('[^\w\s]','')
df
```

```
#removal of stopwords
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['Removal of stopwords'] = df['Punctuation removal'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df
```

```
#removal of frequent words
frequent_words = pd.Series(' '.join(df['Removal of stopwords']).split()).value_counts()[:10]
print(frequent_words)
frequent_words_list = list(frequent_words.index)
df['frequent words removal'] = df['Removal of stopwords'].apply(lambda x: " ".join(x for x in x.split() if x not in frequent_words_list))
df
```

```
#removal of rare words
rarewords = pd.Series(' '.join(df['Removal of stopwords']).split()).value_counts()[-10:]
print(rarewords)
rarewords = list(rarewords.index)
df['Rare Words removal'] = df['frequent words removal'].apply(lambda x: " ".join(x for x in x.split() if x not in rarewords))
df
```

```
#spelling correction
from textblob import TextBlob
df['Correction of spelling'] = df['Rare Words removal'].apply(lambda x: str(TextBlob(x).correct()))
df
```

```
import nltk
nltk.download('punkt')
df['Tokenization'] = df['Correction of spelling'].apply(lambda x: TextBlob(x).words)
df
```

```
#stemming
from nltk.stem import PorterStemmer
port = PorterStemmer()
df['Stemming'] = df['Tokenization'].apply(lambda x: " ".join([port.stem(word) for word in x]))
df
```

```
#lemmatization
from textblob import Word
import nltk
nltk.download('wordnet')
df['Lemmatization'] = df['Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df
```

```
#1.3 saving to csv file
df.to_csv('final_data.csv',index=False)
```

```
#1.4
term_freq = (df['Lemmatization']).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
term_freq.columns = ['words','term_frequency']
term_freq
```

```
import itertools
import collections
from nltk import ngrams
terms = []
for i in df['Lemmatization']:
  terms.append(nltk.tokenize.word_tokenize(i))
text = [i for i in terms if i != []]
value = list(itertools.chain.from_iterable(text))
gram = ngrams(value, 1)
print("1 grams")
print(collections.Counter(gram).most_common(10))
print("Bi grams")
freq = nltk.FreqDist(nltk.bigrams(value))
print(freq.most_common(10))
print("Tri grams")
freq = nltk.FreqDist(nltk.trigrams(value))
print(freq.most_common(10))
```
